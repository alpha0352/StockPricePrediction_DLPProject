```python
import requests
import pandas as pd
import warnings
from bs4 import BeautifulSoup
from datetime import datetime
warnings.filterwarnings('ignore')
# URL of the page to scrape
url = "https://fortune.com/sitemap/2017/12/"
# Set a user-agent header to make the request look like it's coming from a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}
# Send an HTTP GET request with the user-agent header
response = requests.get(url, headers=headers)
# print(response.status_code)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    href_list = [url + element['href'] for element in soup.select('ol.sitemap-grid-months a')]
    # href_list =['https://fortune.com/sitemap/2017/12/1/']
else:
    print("Failed to retrieve the page. Status code:", response.status_code)
print(href_list)
```

```python
# href_list = ['https://fortune.com/sitemap/2018/04/20/', 'https://fortune.com/sitemap/2018/04/21/', 'https://fortune.com/sitemap/2018/04/22/', 'https://fortune.com/sitemap/2018/04/23/', 'https://fortune.com/sitemap/2018/04/24/', 'https://fortune.com/sitemap/2018/04/25/', 'https://fortune.com/sitemap/2018/04/26/', 'https://fortune.com/sitemap/2018/04/27/', 'https://fortune.com/sitemap/2018/04/28/', 'https://fortune.com/sitemap/2018/04/29/', 'https://fortune.com/sitemap/2018/04/30/']
Fortune_dataframe = pd.DataFrame(columns=['date','category', 'article title'])
acceptable_category = [' Finance ']
for href in href_list:
    formatted_date = datetime.strptime(href[28:len(href)-1], "%Y/%m/%d").strftime("%d/%m/%Y")
    date_response = requests.get(href,headers=headers)
    if date_response.status_code == 200:
        article_soup = BeautifulSoup(date_response.text,'html.parser')
        # print(date_response.text)
        article_links = [element['href'] for element in article_soup.select('.sitemap-grid-articles a')]
    for articles in article_links:
        # print(articles)
        article_page = requests.get(articles,headers=headers)
        if article_page.status_code == 200:
            article_soup = BeautifulSoup(article_page.text,'html.parser')
            # print(article_cat.text)
            try:
                article_cat = article_soup.find('div', class_='sc-2c664cc8-1').find('a')
            except Exception:
                article_cat = None
            if article_cat:
                # print(article_cat.text)
                if article_cat.text in acceptable_category:
                    print(articles)
                    # try:
                    #     article_title = article_soup.find('h1', class_='sc-2fb31a5c-2').text
                    # except AttributeError:
                    #     article_title = article_soup.find('h1', class_='sc-7d19b5f1-3').text
                    article_title = article_soup.select('#content h1')[0].text
                    print(article_title)
                    Fortune_dataframe = Fortune_dataframe.append({'date': formatted_date,'category':article_cat.text,'article title': article_title}, ignore_index=True)
existing_df = pd.read_csv('Fortune.csv')
combined_pd = pd.concat([existing_df, Fortune_dataframe], ignore_index=True)
combined_pd.to_csv('Fortune.csv',index=False)
# Fortune_dataframe.to_csv('Fortune_new.csv')

```
