```python
import requests
import pandas as pd
import warnings
from bs4 import BeautifulSoup
from datetime import datetime
warnings.filterwarnings('ignore')
# URL of the page to scrape
url = "https://www.wsj.com/news/archive/2018/may"

# Set a user-agent header to make the request look like it's coming from a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}

# Send an HTTP GET request with the user-agent header
response = requests.get(url, headers=headers)
# print(response.status_code)

# Check if the request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    articles = soup.find_all(class_ = "WSJTheme--day-link--19pByDpZ")
    # print(articles)
    
    # href_list = ['/news/archive/2017/12/01']
    href_list = [element['href'] for element in articles]
    # for href in href_list:
        # print("href:", href)

else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```

```python

WSJ_dataframe = pd.DataFrame(columns=['date','category', 'article title'])
pages = ['1','2','3','4','5','6']

acceptable_category = [' Financial Regulation',' Central Banks',' Financial Regulation', ' Markets', ' Business',' U.S. Markets']
for href in href_list:
    date = "/".join(href.split("/")[3:])  # Join parts from index 3 onwards
    parsed_date = datetime.strptime(date, "%Y/%m/%d")
    formatted_date = parsed_date.strftime("%Y-%m-%d")
    for page_no in pages:
        date_response = requests.get('https://www.wsj.com'+href+'?'+page_no,headers=headers)
        # print(date_response)
        if date_response.status_code == 200:
            # print('hello world')
            article_soup = BeautifulSoup(date_response.text,'html.parser')
            articles = article_soup.find_all('div',class_ = "WSJTheme--overflow-hidden--qJmlzHgO")
            # print(articles)
            for article in articles:
                article_category = article.find('div',class_= "WSJTheme--articleType--34Gt-vdG").text
                # print(article_category)
                if article_category in acceptable_category:
                    article_title = article.find('span', class_='WSJTheme--headlineText--He1ANr9C').text
                    WSJ_dataframe = WSJ_dataframe.append({'date': formatted_date,'category':article_category,'article title': article_title}, ignore_index=True)
                    # print(article_title)
# print(WSJ_dataframe)
# csv_file_path = 'WSJ.csv'

# # Save the DataFrame to a CSV file
# WSJ_dataframe.to_csv(csv_file_path, index=False)
existing_df = pd.read_csv('WSJ.csv')
combined_pd = pd.concat([existing_df, WSJ_dataframe], ignore_index=True)
combined_pd.to_csv('WSJ.csv',index=False)
```
