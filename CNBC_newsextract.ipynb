```python
import requests
import pandas as pd
import warnings
from bs4 import BeautifulSoup
from datetime import datetime
warnings.filterwarnings('ignore')
# URL of the page to scrape
url = "https://www.cnbc.com/site-map/articles/2018/May/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    articles_date = soup.select('.SiteMapDay-fullDate a')
    # print(articles_date)
    # href_list = ['https://www.cnbc.com/site-map/articles/2018/February/7/']
    href_list = ['https:' + element['href'] for element in articles_date]
else:
    print("Failed to retrieve the page. Status code:", response.status_code)
print(href_list)
```

```python
# href_list = ['https://www.cnbc.com/site-map/articles/2018/March/22/','https://www.cnbc.com/site-map/articles/2018/March/23/', 'https://www.cnbc.com/site-map/articles/2018/March/24/', 'https://www.cnbc.com/site-map/articles/2018/March/25/', 'https://www.cnbc.com/site-map/articles/2018/March/26/', 'https://www.cnbc.com/site-map/articles/2018/March/27/', 'https://www.cnbc.com/site-map/articles/2018/March/28/', 'https://www.cnbc.com/site-map/articles/2018/March/29/', 'https://www.cnbc.com/site-map/articles/2018/March/30/', 'https://www.cnbc.com/site-map/articles/2018/March/31/']
CNBC_dataframe = pd.DataFrame(columns=['date','category', 'article title'])
acceptable_category = ['Finance', 'Markets','Banks']
for href in href_list:
    formatted_date = datetime.strptime(href[39:len(href)-1], "%Y/%B/%d").strftime("%d/%m/%Y")
    date_response = requests.get(href,headers=headers)
    if date_response.status_code == 200:
        article_soup = BeautifulSoup(date_response.text,'html.parser')
        # print(date_response.text)
        article_links = [element['href'] for element in article_soup.select('.SiteMapArticleList-articleData a')]
        print(article_links)

    for articles in article_links:
        print(articles)
        article_page = requests.get(articles,headers=headers)
        # print(article_page.status_code)
        if article_page.status_code == 200:
            article_soup = BeautifulSoup(article_page.text,'html.parser')
            try:
                article_cat = article_soup.select('#main-article-header a')[0].text
            except IndexError:
                article_cat = None
            print(article_cat)
            if article_cat in acceptable_category:
                article_title = article_soup.select('#main-article-header h1')[0].text
                print(article_title)
                CNBC_dataframe = CNBC_dataframe.append({'date': formatted_date,'category':article_cat,'article title': article_title}, ignore_index=True)
existing_df = pd.read_csv('CNBC.csv')
combined_pd = pd.concat([existing_df, CNBC_dataframe], ignore_index=True)
combined_pd.to_csv('CNBC.csv',index=False)
print(CNBC_dataframe)
```
